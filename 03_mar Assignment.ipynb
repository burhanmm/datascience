# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Sure, here are five plots that can be created using the Seaborn library and their uses:

1. Scatter plot: A scatter plot is used to visualize the relationship between two variables, where one variable is plotted on the x-axis, and the other variable is plotted on the y-axis. It can help identify trends, clusters, and outliers in the data.

2. Line plot: A line plot is used to visualize the trend of a variable over time or across different categories. It is useful for displaying changes and patterns in the data.

3. Histogram: A histogram is used to visualize the distribution of a single variable. It displays the frequencies of different values or ranges of values in the data.

4. Bar plot: A bar plot is used to compare different categories or groups based on a single variable. It is useful for displaying the differences in values among different groups.

5. Heatmap: A heatmap is used to visualize the correlation between different variables in a dataset. It displays the values in a color-coded format, where darker colors indicate higher correlation values. It is useful for identifying relationships and patterns between variables.

# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns

fmri = sns.load_dataset('fmri')
sns.lineplot(x = 'timepoint',y = 'signal', hue="event", style="region", data = fmri)
```

# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')


sns.boxenplot(x = 'pclass', y = 'age',data = titanic)

sns.boxenplot(x = 'pclass', y = 'fare',data = titanic)
```

# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')

sns.displot(x = 'price', hue = 'cut', data = diamonds)
```

# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns

iris = sns.load_dataset('iris')

sns.pairplot(iris,hue = 'species')
```

# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

flights = sns.load_dataset('flights')

sns.heatmap(flights.pivot("month", "year", "passengers"), cmap="YlOrRd")
```
