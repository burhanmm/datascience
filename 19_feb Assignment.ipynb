# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight web framework written in Python. It is designed to be simple and easy to use, while still being powerful enough to build complex web applications. Flask is classified as a micro-framework because it does not require particular tools or libraries, but it is still capable of handling the demands of a web application.

Some of the advantages of Flask Framework are:

Lightweight: Flask is a lightweight framework that does not have any external dependencies, making it easy to install and use.

Flexible: Flask is very flexible and can be used to build a wide range of web applications, from small personal websites to large enterprise systems.

Simple to use: Flask's simple and intuitive API makes it easy to learn and use. It also has a comprehensive documentation that makes it easy for developers to get started.

Extensible: Flask is extensible, allowing developers to add functionality through third-party libraries and extensions.

Modular: Flask is modular, which means that developers can use only the parts of the framework that they need for their specific application, making it easy to customize.

Testability: Flask makes it easy to write test cases, making it easy to test applications during development.

Scalability: Flask can handle high traffic applications and can be scaled up or down depending on the needs of the application.

# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
# pip install flask

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"    

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

# Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is like a map that connects URLs to specific functions in a web application. We use app routes to define the different URLs that users can access and the functions that should be executed when those URLs are accessed. This helps to create a clear structure for web applications, makes it easy to implement RESTful APIs, and allows developers to organize and maintain the code of the application.

# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It takes the endpoint name as its first argument and any number of keyword arguments, where the keyword corresponds to the variable part of the URL rule.

```python
# pip install flask

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

with app.test_request_context():
    print(url_for('profile', username='burhan'))
    
    # Output: /user/burhan
    
if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
