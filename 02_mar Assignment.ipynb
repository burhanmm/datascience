# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a Python data visualization library that provides a wide range of tools for creating high-quality graphs, charts, and other visual representations of data. It is widely used for scientific computing, data analysis, and machine learning applications.

Matplotlib is used for visualizing data in Python. It can be used to create a wide range of plots, including line plots, scatter plots, bar charts, histograms, and pie charts. Matplotlib provides a simple and intuitive interface that makes it easy to create and customize these plots.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line plot: A line plot is used to visualize the trend of data over time. It is created by connecting a series of data points with a straight line.

2. Scatter plot: A scatter plot is used to visualize the relationship between two variables. It is created by plotting each data point as a point on a graph.

3. Bar chart: A bar chart is used to compare the values of different categories. It is created by plotting bars of different heights to represent the values of each category.

4. Histogram: A histogram is used to visualize the distribution of a set of data. It is created by dividing the data into bins and plotting the number of data points in each bin.

5. Pie chart: A pie chart is used to visualize the proportion of each category in a dataset. It is created by dividing a circle into sectors, with each sector representing a different category and the size of the sector proportional to the value of the category.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter plot is visual representtion of data of x and y axis in dotted point in 2 dimentional manner.

A scatter plot is a type of graph that displays the relationship between two numerical variables. It is used to see if there is a correlation or pattern between the two variables. In a scatter plot, each data point is represented as a dot, with the position of the dot determined by the values of the two variables being compared. The horizontal axis represents one variable and the vertical axis represents the other variable. Scatter plots are commonly used in data analysis, scientific research, and business analytics to identify trends and relationships between variables. They are also useful for detecting outliers or unusual observations that do not follow the overall pattern of the data.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
 
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("x and y  scatter graph title")
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

The subplot() function in Matplotlib is used to create multiple plots in a single figure. This function divides the figure into a grid of subplots and specifies the position of each subplot using three arguments: the number of rows, the number of columns, and the index of the subplot.

```python
import matplotlib.pyplot as plt
import numpy as np

# define the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# create a 2x2 grid of subplots and set the size of the figure
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# plot the first line plot in the top left subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# plot the second line plot in the top right subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# plot the third line plot in the bottom left subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# plot the fourth line plot in the bottom right subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# add a title for the entire figure
fig.suptitle('Four Line Plots')

# display the figure
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of graph that displays categorical data with rectangular bars, where the length or height of each bar represents the frequency, count, or proportion of data in each category. Bar plots are used to compare the values of different categories or to show the distribution of data within each category. They are commonly used in data analysis, scientific research, and business analytics to present results or findings to an audience.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit, color = 'r', alpha = 0.5)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Horizontal Bar Plot Title")
plt.grid()
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot is a type of graph that displays the distribution of numerical data through five summary statistics: minimum, maximum, median, first quartile (25th percentile), and third quartile (75th percentile). It is used to identify outliers, compare the distribution of data between different groups, and visualize the spread and skewness of the data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

import matplotlib.pyplot as plt
import numpy as np

# define the data for the box plot
data = [box1, box2]

# create a box plot
plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
