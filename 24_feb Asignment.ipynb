# Q2. Given a Pandas DataFrame df with columns 'A', 'B', and 'C', write a Python function to re-index the DataFrame with a new index that starts from 1 and increments by 2 for each row.

```python
import pandas as pd

def reindex_df(df):
    # Get the number of rows in the DataFrame
    num_rows = len(df)

    # Create a new index that starts from 1 and increments by 2 for each row
    new_index = range(1, 2*num_rows, 2)

    # Set the new index of the DataFrame
    df.index = new_index

    # Return the updated DataFrame
    return df
```

```python
# Create a sample DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

# Call the function to re-index the DataFrame
df = reindex_df(df)

# Print the updated DataFrame
print(df)
```

# Q4. Given a Pandas DataFrame df with a column 'Text', write a Python function to create a new column 'Word_Count' that contains the number of words in each row of the 'Text' column.

```python
import pandas as pd

# Create a sample DataFrame
df1 = pd.DataFrame({'Text': ['I am burhan', 'Student of Data Science', 'Batch of 2023']})


def add_word_count_column(df):
    # Create a new column 'Word_Count' and initialize with 0
    df['Word_Count'] = 0

    # Loop through each row of the DataFrame
    for i in range(len(df)):
        # Split the text in the 'Text' column into words
        words = df.loc[i, 'Text'].split()

        # Count the number of words and store in the 'Word_Count' column
        df.loc[i, 'Word_Count'] = len(words)

    # Return the updated DataFrame
    return df

df2 = add_word_count_column(df1)

print(df2)
```

# Q5. How are DataFrame.size() and DataFrame.shape() different?

Both DataFrame.size() and DataFrame.shape() are methods in the pandas library that can be called on a DataFrame to get information about its dimensions. However, they return different types of information.

DataFrame.size() returns the total number of elements in the DataFrame, which is equal to the number of rows multiplied by the number of columns. For example, if a DataFrame has 5 rows and 3 columns, DataFrame.size() will return 15.

On the other hand, DataFrame.shape() returns a tuple that contains the number of rows and columns in the DataFrame. For example, if a DataFrame has 5 rows and 3 columns, DataFrame.shape() will return the tuple (5, 3).

To summarize:

DataFrame.size() returns the total number of elements in the DataFrame

DataFrame.shape() returns a tuple containing the number of rows and columns in the DataFrame

# Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days for each row in the DataFrame. The moving average should be calculated using a window of size 7 and should include the current day.

```python
import pandas as pd

# Creating a sample DataFrame
df = pd.DataFrame({'Sales': [10, 20, 30, 40, 50, 60, 70, 80, 90],
                   'Date': ['2022-03-01', '2022-03-02', '2022-03-03', '2022-03-04', '2022-03-05', '2022-03-06', '2022-03-07', '2022-03-08', '2022-03-09']})

def moving_average(df):
    df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
    return df

# Applying the moving_average function
df = moving_average(df)

# Displaying the updated DataFrame
df
```

```python
import pandas as pd

def add_weekday_column(df):
    df['Weekday'] = pd.to_datetime(df['Date']).dt.day_name()
    return df

# Creating a sample DataFrame
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']})

# Applying the add_weekday_column function
df = add_weekday_column(df)

# Displaying the updated DataFrame
print(df)
```

# Q12. Given a Pandas DataFrame df with a column 'Date' that contains timestamps, write a Python function to select all rows where the date is between '2023-01-01' and '2023-01-31'.

```python
import pandas as pd

def select_january_dates(df):
    df['Date'] = pd.to_datetime(df['Date'])
    mask = (df['Date'] >= '2023-01-01') & (df['Date'] <= '2023-01-31')
    result = df.loc[mask]
    return result
```

```python
# Creating a sample DataFrame
df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-05', '2023-01-15', '2023-01-20', '2023-02-01'], 'Sales': [100, 200, 300, 400, 500]})

# Applying the select_january_dates function
result = select_january_dates(df)

# Displaying the resulting DataFrame
print(result)
```
